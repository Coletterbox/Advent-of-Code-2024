```python
import re
```

```python
filename = "Day 3 Input.txt"
```

```python
def parse_input(filename):
    input_file = open(filename, "r")
    input_list = input_file.readlines()
    
    s = ""
    output = s.join(input_list)
    output = output.replace("\n", "")
    
    return output
```

```python
# parse_input(filename)
```

```python
def find_mul(input_str):
    commands = re.findall("mul\(+[000-999]+,+[000-999]+\)", input_str)
    
    return commands
```

```python
# find_mul(parse_input(filename))
```

```python
def multiply_numbers(commands):
    total = 0
    
    for command in commands:
        numbers0 = command.replace("mul(", "")
        numbers = numbers0.replace(")", "")
        number_pair = numbers.split(",")
        product = int(number_pair[0]) * int(number_pair[1])        
#         print(number_pair, product)
        
        total += product
        
    return total
```

# Part 1

```python
multiply_numbers(find_mul(parse_input(filename)))
```

```python
def find_all_commands(input_str):
    commands = re.findall("do\(\)|don\'t\(\)|mul\(+[000-999]+,+[000-999]+\)", input_str)
    
    return commands
```

```python
# find_all_commands(parse_input(filename))
```

```python
def find_recent(commands): 
    true_or_false = []
    
    i = 0
    while i < len(commands):
        ii = i
        do = True
        while ii >= 0:
            if commands[ii] == "do()":
                do = True
                break
            elif commands[ii] == "don\'t()":
                do = False
                break
            ii -= 1
        print(do, commands[i])
        true_or_false.append(str(do) + " " + commands[i])
        i += 1
    
    return true_or_false
```

```python
find_recent(find_all_commands(parse_input(filename)))
```

```python
def find_new_total(commands):
    total = 0
    
    true_or_false = find_recent(commands)
    
    for command in true_or_false:
        command_split = command.split(" ")
        print("command_split", command_split)
        if command_split[0] == "True" and command_split[1] != "do()":
            print("command", command)
            numbers0 = command.replace("mul(", "")
            numbers = numbers0.replace(")", "")
            number_pair = numbers.split(",")
            product = int(number_pair[0]) * int(number_pair[1])        
            total += product
        else:
            total += 0
            
    return total
        
```
